```python
from matplotlib import pyplot as plt
#from matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.arange(1,10,0.1)
x
```

```python
y =5*x + 8
y
```

```python
plt.plot(x,y)
plt.title('y=mx+c')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

```python
y2=8*x+10
y2
```

```python
#many graph in one figure
plt.plot(x,y,color='g')
plt.plot(x,y2,color='r')
plt.legend(['y','y2'])
plt.title('y=mx+c')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

```python
# use of subplot to have separate graph
plt.subplot(1,2,1)
plt.plot(x,y,color='g')
#subplot(no.of rows,no.of columns,which graph to activate)
plt.subplot(1,2,2)
plt.plot(x,y2,color='r')
plt.show()
```

```python
x = [2,2,2,2,2,2,2,3,3,3,3,3,3,5,5,5,5,5,7,7,7,7,6,7,7,7,6,6,8,8,8,8,9,9,9,9,9,9,9,9]
#histogram is for continious data (frequency distributiion)
plt.hist(x,color='g')
plt.show()
```

```python
## boxplot for distribution (description of grouped data)
x=[1,2,4,2]
y=[14,9,9,2]
z=[4,5,7,8]
a33=list([x,y,z])
a33
plt.boxplot(a33)
plt.show()
```

```python
#violin plot for population density
plt.violinplot(a33,showmeans=True)
plt.show()
```

```python
#pie plot info
x2221=['Apple','Orange','Grapes']
y2221=[20,16,80]
plt.pie(y2221,labels=x2221,autopct='%0.2f%%',explode=(0,0.2,0),shadow=True)
plt.show()
```

```python
# donut plot
plt.pie(y2221,labels=x2221,radius=1,autopct='%0.2f%%',explode=(0.4,0,0))
plt.pie([3],radius=0.4,colors='w')
plt.show()
```
